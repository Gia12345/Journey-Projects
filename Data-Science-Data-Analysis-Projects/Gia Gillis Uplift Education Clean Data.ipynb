```python
def parse_columns(df):
    for i in df.columns:
        df[i]=df[i].apply(lambda x: parse_data(x, i))
    return df

def parse_data(data, column_name):
    value=data
    if value is not None and isinstance(value, str) and value is not '':
        if 'Days' in value:
            value=int(value.replace('Days', ''))
        elif 'hours' in value:
            value=float(value.replace('hours', ''))
        elif value[len(value)-1]=='%':
            value = float(value.replace('%',''))
        else: 
            value=int(value)
#        elif 'Percent' in column_name:
#            value = float(value)/100
    return value
```
