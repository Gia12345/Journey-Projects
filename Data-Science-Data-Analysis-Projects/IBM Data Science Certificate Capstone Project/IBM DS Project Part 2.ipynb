##  Neighborhoods in Toronto, Canada

Import libraries.

```python
import numpy as np
import pandas as pd
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

Get information from Wikipedia page.

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
response = requests.get(url)
response
```

Use BeautifulSoup to parse web page.

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

Get table from web page.

```python
table=soup.find_all('tr')
```

```python
rows=soup.find_all('td')
```

```python
l=[]
for each in rows:
    try:
        l.append(each.contents[0].contents[0])
        print(each.contents[0].contents[0])
    except:
        l.append(each.contents[0])
        print(each.contents[0])    
```

```python
l
```

Divide table into rows.

```python
def divide_chunks(l, n): 
    # looping till length l 
    for i in range(0, len(l), n):  
        yield l[i:i + n] 
n = 3
x = list(divide_chunks(l, n)) 
x
```

Create dataframe from list.

```python
df=pd.DataFrame(x)
df.columns=['PostalCode', 'Borough', 'Neighborhood']
```

Drop rows that contain 'Not assigned' as Borough.

```python
df=df.astype('str')
```

```python
df = df[df['Borough'] != 'Not assigned']
```

For Neighborhood with 'Not assigned', change Neighborhood name to Borough name.
Remove '\n' from Neighborhood.

```python
for index, row in df.iterrows(): 
    if 'Not assigned' in row['Neighborhood']:
        row['Neighborhood']=row['Borough']
    if '\n' in row['Neighborhood']:
        row['Neighborhood']=row['Neighborhood'][:-1]
```

Set multi index and combine Neighborhoods with the same PostalCode and Borough.

```python
df.set_index(['PostalCode','Borough'],inplace=True)
grouped = df.groupby(level=['PostalCode','Borough'], sort=False).agg( ','.join)
grouped
```

```python
grouped.reset_index(inplace=True)
```
