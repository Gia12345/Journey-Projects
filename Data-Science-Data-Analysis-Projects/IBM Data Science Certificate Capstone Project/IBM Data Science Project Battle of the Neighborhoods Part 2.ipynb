##  Walt Disney World Resort Areas Part 2

Get latitude and longitude information and join with csv file of resort hotel information created in previous notebook.  Save combined info to csv file.

Import libraries.

```python
import pandas as pd
import numpy as np
```

Read csv file with Toronto, Canada postal codes with latitudes and longitudes.

```python
df=pd.read_csv('wdw_resorts_lat_long.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

Set index to 'PostalCode' to match neighborhoods file.

```python
df=df.set_index('Resort')
```

Read csv file with Resort data that was created in previous Jupyter Notebook.

```python
resorts=pd.read_csv('wdw_resorts.csv')
```

```python
resorts.set_index('Resort', inplace=True)
resorts
```

Join neighborhoods dataframe and latitude and longitude dataframe.

```python
joined=resorts.join(df, on='Resort', how='right')
joined
```

```python
joined=pd.concat([resorts, df], axis=1)
joined
```

Save joined dataframe to csv file.

```python
joined.to_csv('wdw_resorts_geo.csv')
```
