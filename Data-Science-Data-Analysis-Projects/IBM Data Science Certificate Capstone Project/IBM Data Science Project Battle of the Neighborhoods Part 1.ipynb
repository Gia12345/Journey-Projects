##  Walt Disney World Resort Areas Part 1

Analyze venues around Walt Disney World Resort hotels.  First, find resort hotel information and save to a csv file.

Import libraries.

```python
import numpy as np
import pandas as pd
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

Get information from Wikipedia page.

```python
url='https://en.wikipedia.org/wiki/Walt_Disney_World'
```

```python
response = requests.get(url)
response
```

Use BeautifulSoup to parse web page.

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

Get rows from web page.

```python
rows=soup.find_all('td')
```

```python
l=[]
for each in rows:
    try:
        l.append(each.contents[0].contents[0])
        print(each.contents[0].contents[0])
    except:
        l.append(each.contents[0])
        print(each.contents[0])    
```

```python
l
```

```python
start=l.index('Resort Area')+1
print('start  ', start)
print(l[start])
end=l.index('Hotel name')
print('end ', end)
print(l[end])
r=l[start:end]
r
```

```python
resorts=[]
#resorts.columns=['Resort Name', 'Date Opened', 'Theme', 'Capacity', 'Resort Area']
i=0
resort_area=''
while i<len(r):
    resorts.append(r[i])
    resorts.append(r[i+1])
    resorts.append(r[i+2])
    if '\n' in r[i+3]:
        resorts.append(r[i+3][:-1])
        i=i+4
    else:
        resorts.append(r[i+3])
        if '\n' in r[i+4]:
            resort_area=r[i+4][:-1]
        else:
            resort_area=r[i+4]
        i=i+5
    resorts.append(resort_area)
resorts
```

Divide table into rows.

```python
def divide_chunks(l, n): 
    # looping till length l 
    for i in range(0, len(l), n):
        yield l[i:i + n] 
n = 5
chunks = list(divide_chunks(resorts, n)) 
chunks
```

Create dataframe from list.

```python
df=pd.DataFrame(chunks)
df.columns=['Resort', 'Opening Date', 'Theme', 'Rooms', 'Resort Area']
df.head()
```

```python
df
```

```python
df=df.astype('str')
df
```

```python
'wiki' in df.iloc[29, 0]
to_drop = df[df.iloc[:, 0].str.contains('wiki')]
```

```python
to_drop
```

```python
df = df.drop(to_drop.index)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.to_csv('wdw_resorts.csv', index=False)
```
